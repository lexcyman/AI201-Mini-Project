```python
import numpy as np

def generate_data(seed=42):
    rng = np.random.default_rng(seed)

    # Generate class -1
    mu1 = np.array([0.0, 0.0])
    om1 = np.eye(2)
    
    class1 = rng.multivariate_normal(mu1, om1, size=100) # class -1
    c1_train = class1[:50]
    c1_test = class1[50:]

    # Generate class +1
    mu2 = np.array([10.0, 10.0])
    om2 = np.eye(2)

    class2 = rng.multivariate_normal(mu2, om2, size=100) # class +1
    c2_train = class2[:50]
    c2_test = class2[50:]

    # Merge training and test sets
    X_train = np.vstack([c1_train, c2_train])  
    y_train = np.hstack([-np.ones(50), np.ones(50)])

    X_test = np.vstack([c1_test, c2_test])
    y_test = np.hstack([-np.ones(50), np.ones(50)])

    return X_train, y_train, X_test, y_test

# X_train, y_train, X_test, y_test = generate_data()
# print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
# print("Unique labels in training set:", np.unique(y_train), "and test set:", np.unique(y_test))
```

```python
class PerceptronPocket:
    def __init__(self, maxitercnt = 10000, seed = 12):
        self.maxitercnt = maxitercnt
        self.seed = seed

        self.w = None
        self.n_v = None
        self.n_w = None
        self.total_itercnt = 0

    def add_bias(self, X):
        N = X.shape[0]
        return np.hstack([X, np.ones((N,1))])

    def sign_with_zero_rule(self, x):
        s = np.sign(x)
        s[s == 0] = 1
        return s
        
    def classify(self, X, y):
        Xb = self.add_bias(X)
        N, d = Xb.shape

        # Set variables to 0
        v = np.zeros(d, dtype=float)
        w = np.zeros(d, dtype=float)
        n_v = 0
        n_w = 0
        itercnt = 0
        rng = np.random.default_rng(self.seed)

        while itercnt <= self.maxitercnt:
            # randomly choose an input pattern xj and associated label yj from the training set
            j = int(rng.integers(0, N))
            xj = Xb[j]
            yj = float(y[j])

            # compute neuron's output given xj
            yj_hat = 1.0 if (v @ xj) >= 0.0 else -1.0

            # If v correctly classifies xj 
            if yj_hat * yj > 0.0:
                n_v += 1          
            else: # Otherwise v misclassifies xj
                if n_v > n_w:
                    w = v.copy()
                    n_w = n_v
                v = v + yj*xj
                n_v = 0
            itercnt += 1

        self.w = w
        self.n_v = n_v
        self.n_w = n_w
        self.total_itercnt = itercnt

        return w

    def predict(self, X):
        Xb = self.add_bias(np.asarray(X, dtype=float))
        test = Xb @ self.w
        return self.sign_with_zero_rule(test).astype(float)

    def evaluate(self, X, y):
        y = np.asarray(y, dtype=float)
        y_pred = self.predict(X)
        err = y - y_pred
        sse = 0.5 * np.sum(err ** 2)
        mse = sse / y.shape[0]

        accuracy = np.mean(y_pred == y)

        TP = np.sum((y_pred == 1) & (y == 1)) 
        TN = np.sum((y_pred == -1) & (y == -1)) 
        FP = np.sum((y_pred == 1) & (y == -1)) 
        FN = np.sum((y_pred == -1) & (y == 1)) 
        
        numerator = (TP * TN) - (FP * FN) 
        denominator = np.sqrt((TP + FP) * (TP + FN) * (TN + FP) * (TN + FN)) 
        mcc = numerator / denominator if denominator != 0 else 0.0
        return mse, accuracy, mcc

# 1. Generate Data
X_train, y_train, X_test, y_test = generate_data(15)

Ptrial = PerceptronPocket()
# Xb_train = Ptrial.add_bias(X_train)
# print(Xb_train.shape) # prints (100, 3)

# 2. Construct Perceptron Classifier [classify()]
pocket_classify = Ptrial.classify(X_train, y_train)
print("Output vector w of the Pocket Algorithm:", pocket_classify)
print("n_w:", Ptrial.n_w, "itercnt_done:", Ptrial.total_itercnt)

# 3. Construct Perceptron Classifier [predict() and evaluate()]
mse, acc, mcc = Ptrial.evaluate(X_test, y_test)
print("MSE=", mse, "| Accuracy = ", acc, "| MCC =", mcc)
```

```python
class AdaBoostPocket:
    def __init__(self, K=10, max_iteration_per_learner=10000, seed=42): 
        self.K = K
        self.max_iter = max_iteration_per_learner
        self.seed = seed
        self.hypothesis = []
        self.alphas = []
        self.rng = np.random.default_rng(seed)

    def adabtrain(self, X, y):
        N = len(y)
        w = np.ones(N) / N
        
        for t in range(self.K):
            # (a) select new training set S_t with replacement according to w_t
            sampling_idx = self.rng.choice(N, size=N, replace=True, p=w)
            XS_t = X[sampling_idx]
            yS_t = y[sampling_idx]

            # (b) train weak learner L on S_t to obtain hypothesis h_t
            learner = PerceptronPocket(maxitercnt=self.max_iter, seed=self.seed+t)
            learner.classify(XS_t, yS_t)

            # (c) compute training error epsilon_t of h_t on S
            predictions = learner.predict(X)
            misclassified = (predictions != y).astype(float)
            epsilon_t = np.sum(w*misclassified)

            # (d) compute coefficient alpha_t
            alpha_t = 0.5 * np.log((1-epsilon_t) / (epsilon_t))

            # (e) update weights
            w *= np.exp(-alpha_t * y * predictions)
            w = w / np.sum(w)

            self.hypothesis.append(learner)
            self.alphas.append(alpha_t)

        return self

    def adabpredict(self, X):
        final_sum = np.zeros(X.shape[0])

        for alpha, learner in zip(self.alphas, self.hypothesis):
            final_sum += alpha*learner.predict(X)

        return np.sign(final_sum)

# X_train, y_train, X_test, y_test = generate_data(15)

# Initialize AdaBoost with Pocket Perceptron [adabtrain()]
ADAtrial = AdaBoostPocket(K=10, max_iteration_per_learner=10000)
ADAtrial.adabtrain(X_train, y_train)

# Evaluate AdaBoost Perceptron [adabpredict()]
y_pred_train = ADAtrial.adabpredict(X_train)
y_pred_test = ADAtrial.adabpredict(X_test)

train_acc = np.mean(y_pred_train == y_train)
test_acc = np.mean(y_pred_test == y_test)

TP = np.sum((y_pred_test == 1) & (y_test == 1)) 
TN = np.sum((y_pred_test == -1) & (y_test == -1)) 
FP = np.sum((y_pred_test == 1) & (y_test == -1)) 
FN = np.sum((y_pred_test == -1) & (y_test == 1)) 

numerator = (TP * TN) - (FP * FN) 
denominator = np.sqrt((TP + FP) * (TP + FN) * (TN + FP) * (TN + FN)) 
mcc = numerator / denominator if denominator != 0 else 0.0

print(f"Training Accuracy: {train_acc:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")
print(f"MCC: {mcc}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn import svm
import time

def load_banana_data(filepath, seed=42):
    rng = np.random.default_rng(seed)
    data = np.loadtxt(filepath, delimiter=',')

    # banana data format: label, feature1, feature2
    y = data[:, 0]
    X = data[:, 1:]

    # shuffle for random sampling
    indices = np.arange(len(X))
    rng.shuffle(indices)

    # sample 400 points for training set and the remaining 4900 for test set
    train_idx = indices[:400]
    test_idx = indices[400:]

    X_train = X[train_idx]
    y_train = y[train_idx]
    X_test = X[test_idx]
    y_test = y[test_idx]

    return X_train, y_train, X_test, y_test

def load_splice_data(filepath, seed=42):
    rng = np.random.default_rng(seed)
    data = np.loadtxt(filepath, delimiter=',')

    # spliced data format: label, feature1, feature2, feature3, ....
    y = data[:, 0]
    X = data[:, 1:]

    # shuffle for random sampling
    indices = np.arange(len(X))
    rng.shuffle(indices)

    # sample 1000 points for training and the remaining 2175 points for testing
    train_idx = indices[:1000]
    test_idx = indices[1000:]

    X_train = X[train_idx]
    y_train = y[train_idx]
    X_test = X[test_idx]
    y_test = y[test_idx]

    return X_train, y_train, X_test, y_test

def run_boosted_perceptron(X_train, y_train, X_test, y_test, K_values, data_name="Dataset"):
    train_accs = []
    test_accs = []
    train_times = []
    test_times = []

    for K in K_values:
        trial = AdaBoostPocket(K=K)

        start_train = time.time()
        trial.adabtrain(X_train, y_train)
        train_time = time.time() - start_train

        start_test = time.time()
        y_train_predict = trial.adabpredict(X_train)
        y_test_predict = trial.adabpredict(X_test)
        test_time = time.time() - start_test

        train_acc = np.sum(y_train == y_train_predict) / len(y_train)
        test_acc = np.sum(y_test == y_test_predict) / len(y_test)

        train_accs.append(train_acc)
        test_accs.append(test_acc)
        train_times.append(train_time)
        test_times.append(test_time)

    return train_accs, test_accs, train_times, test_times

def run_svm_classifier(X_train, y_train, X_test, y_test, kernel='rbf', C=1.0, gamma='scale', data_name="Dataset"):
    svm_clf = svm.SVC(kernel=kernel, C=C, gamma=gamma)

    start_train = time.time()
    svm_clf.fit(X_train, y_train)
    train_time = time.time() - start_train

    start_test = time.time()
    y_train_predict = svm_clf.predict(X_train)
    y_test_predict = svm_clf.predict(X_test)
    test_time = time.time() - start_test

    train_acc = np.sum(y_train == y_train_predict) / len(y_train)
    test_acc = np.sum(y_test == y_test_predict) / len(y_test)

    return train_acc, test_acc, train_time, test_time

def plot_accuracies(K_list, train_accuracies, test_accuracies, data):
    plt.figure(figsize=(8,5))
    plt.plot(K_list, train_accuracies, marker= 'o', label="Training Accuracy")
    plt.plot(K_list, test_accuracies, marker='x', label="Test Accuracy")

    plt.xlabel("Number of Learners (K)")
    plt.ylabel("Accuracy")
    plt.title(f"Boosted Perceptron (Pocket Algo) Performance with {data} Dataset")

    plt.legend()
    plt.grid(True)
    plt.savefig(f"K_vs_accuracy_{data}.png", dpi=300, bbox_inches='tight')
    plt.show()

def plot_kernels(results, data):
    df = pd.DataFrame(results, columns=['kernel', 'C', 'gamma', 'train_acc', 'test_acc', 'train_time', 'test_time'])

    for kernel in df['kernel'].unique():
        subset = df[df['kernel'] == kernel].pivot(index='gamma',columns='C',values='test_acc')

        plt.figure(figsize=(6,4))
        sns.heatmap(subset, annot=True, cmap="viridis")

        plt.title(f"SVM {kernel} Kernel Accuracies on {data} Dataset")
        plt.xlabel("C")
        plt.ylabel("Gamma")
        
        plt.savefig(f"SVM_kernel_config_{kernel}_{data}", dpi=300, bbox_inches='tight')
        plt.show()
        
def run_best_svm(X_train, y_train, X_test, y_test, kernel_list, C_list, gamma_list, data_name):
    best_acc = -1
    best_params = None
    results = []

    for kernel in kernel_list:
        for C in C_list:
            for gamma in gamma_list:
                train_acc, test_acc, train_time, test_time = run_svm_classifier(X_train, y_train, X_test, y_test, kernel=kernel, C=C, gamma=gamma)
                results.append((kernel, C, gamma, train_acc, test_acc, train_time, test_time))
                if test_acc > best_acc:
                    best_acc = test_acc
                    best_params = (kernel, C, gamma, test_acc)

    print("="*70)
    print(f"SVM Configuration Results for {data_name} Dataset")
    print(f"{'Kernel':<10}{'C':>8}{'Gamma':>12}{'Train Accuracy':>14}{'Test Accuracy':>14}{'Training Time (s)':>18}{'Test Time (s)':>18}")
    for kernel, C, gamma, train_acc, test_acc, train_time, test_time in results:
        print(f"{kernel:<10}{C:>8}{str(gamma):>12}{train_acc:14.4f}{test_acc:14.4f}{train_time:18.4f}{test_time:18.4f}")
    print("-"*90)
    
    print(f"Best SVM Parameters for {data_name} Dataset: Kernel={best_params[0]}, C={best_params[1]}, Gamma={best_params[2]}, Test Accuracy={best_acc:.4f}")
    return best_params, results
    
def compare_models(abp_train_acc, abp_test_acc, abp_train_time, abp_test_time, svm_train_acc, svm_test_acc, svm_train_time, svm_test_time, data_name="Dataset"):
    # Best AdaBoost Perceptron (ABP) based on test accuracy
    best_index = np.argmax(abp_test_acc)
    best_abp_train_acc = abp_train_acc[best_index]
    best_abp_test_acc = abp_test_acc[best_index]
    best_abp_train_time = abp_train_time[best_index]
    best_abp_test_time = abp_test_time[best_index]

    print("="*70)
    print(f"Performance Evaluation on {data_name} Dataset")
    print(f"{'Model':<25}{'Train Accuracy':>12}{'Test Accuracy':>12}{'Train Time (s)':>16}{'Test Time (s)':>16}")
    print(f"{'Boosted Perceptron':<25}{best_abp_train_acc:12.4f}{best_abp_test_acc:12.4f}{best_abp_train_time:16.4f}{best_abp_test_time:16.4f}")   
    print(f"{'SVM':<25}{svm_train_acc:12.4f}{svm_test_acc:12.4f}{svm_train_time:16.4f}{svm_test_time:16.4f}")
```

```python
# Example: load banana data
X_train_banana, y_train_banana, X_test_banana, y_test_banana = load_banana_data('banana_data.csv')

# Example: load spliced data
X_train_splice, y_train_splice, X_test_splice, y_test_splice = load_splice_data('splice_data.csv')

# Hyperparameter ranges
K_values = list(range(10, 1001, 25))
kernel_list = ['linear', 'rbf', 'poly']
C_list = [0.1, 1, 10]
gamma_list = ['scale', 0.1, 1]

# SVM: Find the kernel and kernel parameters to obtain the best performance
best_parameters_banana, banana_results = run_best_svm(X_train_banana, y_train_banana, X_test_banana, y_test_banana, kernel_list, C_list, gamma_list, data_name="Banana")
best_parameters_splice, splice_results = run_best_svm(X_train_splice, y_train_splice, X_test_splice, y_test_splice, kernel_list, C_list, gamma_list, data_name="Spliced")

kernel_banana, C_banana, gamma_banana, _ = best_parameters_banana
kernel_splice, C_splice, gamma_splice, _ = best_parameters_splice

# For Banana data (abp = AdaBoost Perceptron, svm = SVM)
abp_train_acc_banana, abp_test_acc_banana, abp_train_time_banana, abp_test_time_banana  = run_boosted_perceptron(X_train_banana, y_train_banana, X_test_banana, y_test_banana, K_values, data_name="Banana")
svm_train_acc_banana, svm_test_acc_banana, svm_train_time_banana, svm_test_time_banana  = run_svm_classifier(X_train_banana, y_train_banana, X_test_banana, y_test_banana, kernel_banana, C_banana, gamma_banana, data_name="Banana")

# For Spliced data
abp_train_acc_splice, abp_test_acc_splice, abp_train_time_splice, abp_test_time_splice = run_boosted_perceptron(X_train_splice, y_train_splice, X_test_splice, y_test_splice, K_values, data_name="Spliced")
svm_train_acc_splice, svm_test_acc_splice, svm_train_time_splice, svm_test_time_splice = run_svm_classifier(X_train_splice, y_train_splice, X_test_splice, y_test_splice, kernel_splice, C_splice, gamma_splice, data_name="Spliced")

# Adaboost: Plot the accuracy vs K curves for both training and test set of Banana and Spliced Datasets
plot_accuracies(K_values, abp_train_acc_banana, abp_test_acc_banana, data="Banana")
plot_accuracies(K_values, abp_train_acc_splice, abp_test_acc_splice, data="Spliced")

# SVM: Find the kernel and kernel parameters to obtain the best performance
plot_kernels(banana_results, data="Banana")
plot_kernels(splice_results, data="Spliced")

# SVM: Compare the performance of SVM with Boosted Perceptrons in terms of test accuracy, training and test speeds
compare_models(abp_train_acc_banana, abp_test_acc_banana, abp_train_time_banana, abp_test_time_banana, 
               svm_train_acc_banana, svm_test_acc_banana, svm_train_time_banana, svm_test_time_banana, data_name="Banana")
compare_models(abp_train_acc_splice, abp_test_acc_splice, abp_train_time_splice, abp_test_time_splice,
               svm_train_acc_splice, svm_test_acc_splice, svm_train_time_splice, svm_test_time_splice, data_name="Spliced")
```

```python
banana_df = pd.DataFrame(banana_results, columns=['kernel', 'C', 'gamma', 'train_acc', 'test_acc', 'train_time', 'test_time'])
splice_df = pd.DataFrame(splice_results, columns=['kernel', 'C', 'gamma', 'train_acc', 'test_acc', 'train_time', 'test_time'])

banana_df.to_csv('svm_kernel_paremeter_settings_results_banana.csv', index=False)
splice_df.to_csv('svm_kernel_paremeter_settings_results_splice.csv', index=False)

performance_data = [
    ['Banana', 'Boosted Perceptron', 
     np.max(abp_train_acc_banana), np.max(abp_test_acc_banana),
     abp_train_time_banana[np.argmax(abp_test_acc_banana)], abp_test_time_banana[np.argmax(abp_test_acc_banana)]],
    ['Banana', 'SVM', svm_train_acc_banana, svm_test_acc_banana, svm_train_time_banana, svm_test_time_banana],
    ['Spliced', 'Boosted Perceptron', 
     np.max(abp_train_acc_splice), np.max(abp_test_acc_splice),
     abp_train_time_splice[np.argmax(abp_test_acc_splice)], abp_test_time_splice[np.argmax(abp_test_acc_splice)]],
    ['Spliced', 'SVM', svm_train_acc_splice, svm_test_acc_splice, svm_train_time_splice, svm_test_time_splice]
]

performance_df = pd.DataFrame(performance_data, 
                              columns=['Dataset', 'Model', 'Train Accuracy', 'Test Accuracy', 'Train Time (s)', 'Test Time (s)'])

performance_df.to_csv('performance_comparison_svm_boosted_perceptron.csv', index=False)
print("All results have been saved.")
```
